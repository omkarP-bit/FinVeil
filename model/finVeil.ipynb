```python
import sys
!{sys.executable} -m pip install ucimlrepo scikit-learn pandas numpy scipy

from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np

statlog_german_credit_data = fetch_ucirepo(id=144)
X_raw = statlog_german_credit_data.data.features
y_raw = statlog_german_credit_data.data.targets

print(X_raw.shape, y_raw.shape)
X_raw.head()
```

```python
y = (y_raw.iloc[:, 0] == 1).astype(int)

categorical_cols = X_raw.select_dtypes(include="object").columns.tolist()
numeric_cols = X_raw.select_dtypes(exclude="object").columns.tolist()
print("Categorical:", categorical_cols)
print("Numeric:", numeric_cols)

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
], remainder="passthrough")

X_encoded = preprocessor.fit_transform(X_raw)
encoded_feature_names = preprocessor.get_feature_names_out()

X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)
print("Full encoded train shape:", X_train_full.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

baseline_model = LogisticRegression(max_iter=2000)
baseline_model.fit(X_train_full, y_train)

baseline_preds = baseline_model.predict(X_test_full)
baseline_accuracy = accuracy_score(y_test, baseline_preds)
print(f"Baseline (full features, true sigmoid) accuracy: {baseline_accuracy:.4f}")
```

```python
N_FEATURES = 6  # tune between 5-8

coef_series = pd.Series(baseline_model.coef_[0], index=encoded_feature_names)
top_features = coef_series.abs().sort_values(ascending=False).head(N_FEATURES).index.tolist()

print("Selected features for Profile struct:")
for f in top_features:
    print(f"  {f}: coef={coef_series[f]:+.4f}")

feature_idx = [list(encoded_feature_names).index(f) for f in top_features]

X_train_small = np.asarray(X_train_full[:, feature_idx].todense()
                            if hasattr(X_train_full, "todense") else X_train_full[:, feature_idx])
X_test_small = np.asarray(X_test_full[:, feature_idx].todense()
                           if hasattr(X_test_full, "todense") else X_test_full[:, feature_idx])
```

```python
small_model = LogisticRegression(max_iter=2000)
small_model.fit(X_train_small, y_train)

small_preds = small_model.predict(X_test_small)
small_accuracy = accuracy_score(y_test, small_preds)
print(f"Small model ({N_FEATURES} features, true sigmoid) accuracy: {small_accuracy:.4f}")
print(f"Accuracy lost by reducing to {N_FEATURES} features: {baseline_accuracy - small_accuracy:+.4f}")

weights = small_model.coef_[0]
bias = small_model.intercept_[0]
print("Weights:", weights)
print("Bias:", bias)
```

```python
from scipy.optimize import curve_fit

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def poly2(z, c0, c1, c2):
    return c0 + c1 * z + c2 * z ** 2

z_train = small_model.decision_function(X_train_small)
z_test = small_model.decision_function(X_test_small)

margin = 0.1 * (z_train.max() - z_train.min())
z_range = np.linspace(z_train.min() - margin, z_train.max() + margin, 400)

coeffs, _ = curve_fit(poly2, z_range, sigmoid(z_range))
c0, c1, c2 = coeffs
print(f"Fitted polynomial coefficients: c0={c0:.6f}, c1={c1:.6f}, c2={c2:.6f}")

poly_probs_test = poly2(z_test, c0, c1, c2)
poly_preds_test = (poly_probs_test >= 0.5).astype(int)
poly_accuracy = accuracy_score(y_test, poly_preds_test)

true_probs_test = sigmoid(z_test)
mean_abs_prob_error = np.abs(poly_probs_test - true_probs_test).mean()

print(f"Small model, true sigmoid accuracy:      {small_accuracy:.4f}")
print(f"Small model, polynomial-activation acc:  {poly_accuracy:.4f}")
print(f"Accuracy delta (poly - sigmoid):          {poly_accuracy - small_accuracy:+.4f}")
print(f"Mean absolute probability error:          {mean_abs_prob_error:.4f}")
```

```python
SCALE = 1000      # fixed-point scale for weights/coeffs
OFFSET = 100_000  # large enough that (value * SCALE) + OFFSET is always >= 0

def quantize_signed(value, scale=SCALE):
    return int(round(value * scale))

def to_unsigned(signed_int, offset=OFFSET):
    v = signed_int + offset
    assert v >= 0, f"OFFSET too small for value {signed_int}; increase OFFSET"
    return v

quantized_weights_signed = [quantize_signed(w) for w in weights]
quantized_bias_signed = quantize_signed(bias)

quantized_weights_unsigned = [to_unsigned(w) for w in quantized_weights_signed]
quantized_bias_unsigned = to_unsigned(quantized_bias_signed)

quantized_c0_signed = quantize_signed(c0)
quantized_c1_signed = quantize_signed(c1)
quantized_c2_signed = quantize_signed(c2)

quantized_c0_unsigned = to_unsigned(quantized_c0_signed)
quantized_c1_unsigned = to_unsigned(quantized_c1_signed)
quantized_c2_unsigned = to_unsigned(quantized_c2_signed)

print("Quantized (unsigned, CoFHE-ready) weights:", quantized_weights_unsigned)
print("Quantized (unsigned) bias:", quantized_bias_unsigned)
print("Quantized (unsigned) poly coeffs [c0, c1, c2]:",
      [quantized_c0_unsigned, quantized_c1_unsigned, quantized_c2_unsigned])
```

```python
N = len(top_features)
FEATURE_SCALE = 100  # separate scale for feature values, tune to their real range

def full_simulation(feature_values):
    z_unsigned_terms = []
    offset_removal = 0
    for qw_unsigned, fv in zip(quantized_weights_unsigned, feature_values):
        fv_q = int(round(fv * FEATURE_SCALE))
        z_unsigned_terms.append(qw_unsigned * fv_q)
        offset_removal += OFFSET * fv_q
    z_unsigned_terms.append(quantized_bias_unsigned)
    offset_removal += OFFSET

    z_unsigned_sum = sum(z_unsigned_terms)
    z_signed_scaled = z_unsigned_sum - offset_removal   # == z * SCALE * FEATURE_SCALE
    z_reconstructed = z_signed_scaled / (SCALE * FEATURE_SCALE)
    return z_reconstructed

sample_idx = 0
z_true = z_test[sample_idx]
z_sim = full_simulation(X_test_small[sample_idx])
print(f"True z (float):        {z_true:.6f}")
print(f"Simulated z (CoFHE):   {z_sim:.6f}")
print(f"Difference:            {abs(z_true - z_sim):.6f}")
```

```python
z_sim_all = np.array([full_simulation(row) for row in X_test_small])

def poly2_quantized_reconstructed(z):
    c0_r = (quantized_c0_unsigned - OFFSET) / SCALE
    c1_r = (quantized_c1_unsigned - OFFSET) / SCALE
    c2_r = (quantized_c2_unsigned - OFFSET) / SCALE
    return c0_r + c1_r * z + c2_r * z ** 2

sim_probs = poly2_quantized_reconstructed(z_sim_all)
sim_preds = (sim_probs >= 0.5).astype(int)
sim_accuracy = accuracy_score(y_test, sim_preds)

print(f"True sigmoid accuracy:                {small_accuracy:.4f}")
print(f"Float polynomial-activation accuracy: {poly_accuracy:.4f}")
print(f"Quantized/offset CoFHE-sim accuracy:  {sim_accuracy:.4f}")
print(f"Quantization error vs float poly:     {sim_accuracy - poly_accuracy:+.4f}")
```

```python
EUINT32_MAX = 2**32 - 1

max_weight_term = max(quantized_weights_unsigned) * (FEATURE_SCALE * max(np.abs(X_test_small).max(), 1))
print(f"Largest individual multiply term (approx): {max_weight_term:,.0f}")
print(f"euint32 max:                                {EUINT32_MAX:,}")
print(f"Headroom OK: {max_weight_term < EUINT32_MAX}")

if max_weight_term >= EUINT32_MAX:
    print("WARNING: reduce OFFSET/SCALE/FEATURE_SCALE, or move to a wider encrypted type.")
```

```python
N = len(top_features)
FEATURE_SCALE = 100

def full_simulation(feature_values):
    z_unsigned_terms = []
    offset_removal = 0
    for qw_unsigned, fv in zip(quantized_weights_unsigned, feature_values):
        fv_q = int(round(fv * FEATURE_SCALE))
        z_unsigned_terms.append(qw_unsigned * fv_q)
        offset_removal += OFFSET * fv_q
    z_unsigned_terms.append(quantized_bias_unsigned)
    offset_removal += OFFSET

    z_unsigned_sum = sum(z_unsigned_terms)
    z_signed_scaled = z_unsigned_sum - offset_removal
    return z_signed_scaled / (SCALE * FEATURE_SCALE)

z_sim_all = np.array([full_simulation(row) for row in X_test_small])

def poly2_quantized_reconstructed(z):
    c0_r = (quantized_c0_unsigned - OFFSET) / SCALE
    c1_r = (quantized_c1_unsigned - OFFSET) / SCALE
    c2_r = (quantized_c2_unsigned - OFFSET) / SCALE
    return c0_r + c1_r * z + c2_r * z ** 2

sim_probs = poly2_quantized_reconstructed(z_sim_all)
sim_preds = (sim_probs >= 0.5).astype(int)
sim_accuracy = accuracy_score(y_test, sim_preds)

print(f"True sigmoid accuracy:                {small_accuracy:.4f}")
print(f"Float polynomial-activation accuracy: {poly_accuracy:.4f}")
print(f"Quantized/offset CoFHE-sim accuracy:  {sim_accuracy:.4f}")
```
